```python
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=83)
```

```python
NB=GaussianNB()
```

```python
from sklearn.model_selection import cross_val_score
```

```python
NB_scores=cross_val_score(NB,x,y,cv=5)
```

```python
print("navies bayes cross validation scores:",NB_scores)
```

```python
print("average navies bayes accurarcy",NB_scores.mean())
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report
```

```python
NB.fit(x_train,y_train)
```

```python
y_pred=NB.predict(x_test)
```

```python
acc=accuracy_score(y_test,y_pred)
cm=confusion_matrix(y_test,y_pred)
cr=classification_report(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
```

```python
rf_scores=cross_val_score(rf,x,y,cv=6)
```

```python
print("RandomForest cross validation score:",rf_scores)
```
